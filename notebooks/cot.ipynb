```python
import pandas as pd
import requests

# ดึงข้อมูลล่าสุด 50 แถวจาก CFTC
url = "https://publicreporting.cftc.gov/resource/jun7-fc8e.json?$limit=50&$order=report_date_as_yyyy_mm_dd DESC"
response = requests.get(url)
df = pd.DataFrame(response.json())

# กรองดูเฉพาะข้อมูลของทองคำ (Gold)
gold_cot = df[df['market_and_exchange_names'].str.contains("GOLD", na=False)]
print(gold_cot[['report_date_as_yyyy_mm_dd', 'noncomm_positions_long_all', 'noncomm_positions_short_all']])
```
